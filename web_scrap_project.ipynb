```python
from bs4 import BeautifulSoup
import requests
# Create lists for all featuress to store the values.
title_data = []
price_data = []
rating_data = []
review_data = []
feature_data = []

# Iterate through each page
for i in range(1,6):
    # Mention the url
    url = f"https://www.flipkart.com/search?q=phone&page={i}"
    # print (url)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers = headers)

    # Break the loop in case there is 403 error
    if response.status_code == 403:
        print("That;s a 403 Error!")
        break

    html_code = response.text
    soup = BeautifulSoup(html_code)

    # Use find_all function to extract the data

    # Product title
    title = soup.find_all("div", attrs = {"class" : "RG5Slk"})
    for t in title:
        title_data.append(t.text)

    # Product ratings
    ratings = soup.find_all("div", attrs = {"class" : "MKiFS6"})
    for r in ratings:
        rating_data.append(r.text)

    # Product price
    prices = soup.find_all("div", attrs = {"class" : "hZ3P6w DeU9vF"})
    for p in prices:
        price_data.append(p.text)

    # Product reviews
    reviews = soup.find_all("span", attrs = {"class" : "PvbNMB"})
    for re in reviews:
        review_data.append(re.text)

    # Product features
    features = soup.find_all("ul", attrs = {"class" : "HwRTzP"})
    for f in features:
        feature_data.append(f.text)


print("Scraping completed")
```

```python
import numpy as np
# Create lists for all featuress to store the values.
title_data = []
price_data = []
rating_data = []
review_data = []
feature_data = []

# Iterate through each page
for i in range(1,43):
    # Mention the url
    url = f"https://www.flipkart.com/search?q=phone&page={i}"
    # print (url)
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
    }

    # Fetch the html contents
    response = requests.get(url, headers = headers)

    # Parse the HTML code
    soup = BeautifulSoup(response.text)

    # Use Containerization
    containers = soup.find_all("a", attrs = {"class" : "k7wcnx"})
    for c in containers:

        # Product title
        title = c.find("div", attrs = {"class" : "RG5Slk"})
        if title is None:
            title_data.append(np.nan)
        else:
            title_data.append(title.text)

        # Product ratings
        rating = c.find("div", attrs = {"class" : "MKiFS6"})
        if rating is None:
            rating_data.append(np.nan)
        else:
            rating_data.append(rating.text)

         # Product price
        price = c.find("div", attrs = {"class" : "hZ3P6w DeU9vF"})
        if price is None:
            price_data.append(np.nan)
        else:
            price_data.append(price.text)

        # Product reviews
        review = c.find("span", attrs = {"class" : "PvbNMB"})
        if review is None:
            review_data.append(np.nan)
        else:
            review_data.append(review.text)

        # Product features
        feature = c.find("ul", attrs = {"class" : "HwRTzP"})
        if feature is None:
            feature_data.append(np.nan)
        else:
            feature_data.append(feature.text)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=20, kde=True)
plt.title("Distribution of Phone Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

**Price Distribution Summary**

*  Most smartphones are priced between ₹5,000 – ₹25,000
*  The data is right-skewed, meaning more low-priced phones and fewer high-priced ones
*  A few premium phones (₹1,00,000+) act as outliers
*  Overall, the market is dominated by budget and mid-range devices  
`Conclusion:` Flipkart listings mainly focus on affordable smartphones, with limited high-end options.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Price', y='Rating', data=df)
plt.title("Price vs Rating")
plt.show()
```

**Price vs Rating Summary**

* Most smartphones have ratings between 4.0 – 4.6, regardless of price
* There is no strong correlation between price and rating
* Some low-priced phones also have high ratings, similar to expensive ones
* Premium phones do not always guarantee better ratings  
`Conclusion:` Higher price does not necessarily mean better user satisfaction; many budget phones perform equally well in ratings.

```python
brand_price = df.groupby('Brand')['Price'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
brand_price.plot(kind='bar')
plt.title("Average Price by Brand")
plt.ylabel("Average Price")
plt.show()
```

**Average Price by Brand Summary**

* Premium brands like Apple have the highest average prices, followed by brands like Xiaomi, OnePlus, and Google
* Mid-range brands such as Samsung, Vivo, and Oppo fall in the moderate price range
* Brands like Redmi, Poco, and Realme offer more budget-friendly smartphones
* Many lesser-known brands have very low average prices, indicating entry-level devices  
`Conclusion:` The market clearly shows a price segmentation, where some brands focus on premium devices while others target budget-conscious consumers.

```python
plt.figure(figsize=(8,5))
sns.countplot(x='RAM', data=df)
plt.title("RAM Distribution")
plt.show()
```

**RAM Distribution Summary**
* Most smartphones have 8 GB RAM, making it the most common configuration
* 4 GB and 6 GB RAM are also widely available, mainly in budget and mid-range phones
* Higher RAM options like 12 GB and above are less common and belong to premium devices
* Very low or unusual values may indicate data inconsistencies or outliers  
`Conclusion:` The market is dominated by 8 GB RAM smartphones, balancing performance and affordability.

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['Price','Rating','RAM','Storage','Rating_Count','Review_Count']].corr(), 
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

* RAM has weak or negative correlation with Price and Rating → RAM alone does not strongly influence price or user satisfaction
* Most other relationships are weak or negligible  
`Conclusion:` Storage and overall popularity (reviews/ratings) influence price more than RAM, while user engagement metrics are strongly interconnected.
